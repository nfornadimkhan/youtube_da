```python
import pandas as pd
import numpy as np
from datetime import datetime

# Read the CSV file
df = pd.read_csv('youtube_data/all_videos_data.csv')

# 1. Remove duplicates
df = df.drop_duplicates()

# 2. Convert published_date to datetime
df['published_date'] = pd.to_datetime(df['published_date'].str.replace('Z', ''))

# 3. Convert numeric columns to appropriate types
numeric_columns = ['duration_seconds', 'view_count', 'like_count', 'comment_count']
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 4. Add human-readable duration column
df['duration_formatted'] = pd.to_datetime(df['duration_seconds'], unit='s').dt.strftime('%H:%M:%S')

# 5. Clean titles (remove extra whitespace and special characters if needed)
df['title'] = df['title'].str.strip()

# 6. Create engagement metrics
df['engagement_rate'] = (df['like_count'] + df['comment_count']) / df['view_count'] * 100

# 7. Drop rows with all missing values
df = df.dropna(how='all')

# 8. Sort by keyword and category
df = df.sort_values(['category','keyword'], ascending=False)

# Save the cleaned dataset
df.to_csv('youtube_data/cleaned_videos_data.csv', index=False)
```

Create a relevance scoring system for the videos based on their titles and the searched keywords:

- Creates a calculate_relevance_score function that uses three metrics:
  - Direct keyword presence (30% weight)
  - Word overlap score (40% weight)
  - Sequence similarity (30% weight)
- Calculates a relevance score (0-100) for each video
  - Adds a relevance category (High/Medium/Low/Not Relevant)
  - Provides option to filter out irrelevant videos

```python
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
import nltk
from difflib import SequenceMatcher
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Download required NLTK data
nltk.download('punkt')

# Initialize BERT model ONCE outside the functions
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

def calculate_base_score(title, keyword):
    """
    Calculate basic relevance score between video title and search keyword
    using multiple metrics
    """
    title = str(title).lower()
    keyword = str(keyword).lower()
    
    # 1. Direct keyword presence (30% weight)
    keyword_present = 1 if keyword in title else 0
    
    # 2. Word overlap score (40% weight)
    title_words = set(word_tokenize(title))
    keyword_words = set(word_tokenize(keyword))
    overlap_score = len(title_words.intersection(keyword_words)) / len(keyword_words) if keyword_words else 0
    
    # 3. Sequence similarity (30% weight)
    sequence_score = SequenceMatcher(None, title, keyword).ratio()
    
    # Combine scores with weights
    final_score = (0.3 * keyword_present + 
                  0.4 * overlap_score + 
                  0.3 * sequence_score) * 100
    
    return round(final_score, 2)

def calculate_relevance_score(row, model=model):
    """
    Enhanced version with semantic similarity using BERT
    """
    title = str(row['title']).lower()
    keyword = str(row['keyword']).lower()
    
    # Get embeddings
    title_embedding = model.encode([title])
    keyword_embedding = model.encode([keyword])
    
    # Calculate semantic similarity
    semantic_score = cosine_similarity(title_embedding, keyword_embedding)[0][0]
    
    # Calculate base score
    base_score = calculate_base_score(title, keyword)
    
    # Combined score (60% base score, 40% semantic similarity)
    final_score = (0.6 * base_score + 0.4 * semantic_score * 100)
    
    return round(final_score, 2)

def main():
    # Read the cleaned dataset
    df = pd.read_csv('youtube_data/cleaned_videos_data.csv')
    
    # Calculate relevance scores
    df['relevance_score'] = df.apply(calculate_relevance_score, axis=1)
    
    # Add relevance category
    def get_relevance_category(score):
        if score >= 45:
            return 'High'
        elif score >= 25:
            return 'Medium'
        elif score >= 10:
            return 'Low'
        else:
            return 'Not Relevant'
    
    df['relevance_category'] = df['relevance_score'].apply(get_relevance_category)
    
    # Filter out irrelevant videos (optional)
    df_relevant = df[df['relevance_score'] > 10]  # Adjust threshold as needed
    
    # Save the updated dataset
    df.to_csv('youtube_data/videos_with_relevance.csv', index=False)
    
    # Print Results Summary
    print("\nRelevance Analysis Results:")
    print("-" * 50)
    
    # Overall statistics
    print("\nOverall Statistics:")
    print(f"Total videos analyzed: {len(df)}")
    print(f"Average relevance score: {df['relevance_score'].mean():.2f}")
    print(f"Median relevance score: {df['relevance_score'].median():.2f}")
    print(f"Score range: {df['relevance_score'].min():.2f} - {df['relevance_score'].max():.2f}")
    
    # Category distribution
    print("\nRelevance Category Distribution:")
    category_counts = df['relevance_category'].value_counts()
    for category, count in category_counts.items():
        percentage = (count / len(df)) * 100
        print(f"{category}: {count} videos ({percentage:.1f}%)")
    
    # Top performing keywords
    print("\nTop 5 Keywords by Average Relevance Score:")
    keyword_performance = df.groupby('keyword')['relevance_score'].agg(['mean', 'count'])
    keyword_performance = keyword_performance[keyword_performance['count'] >= 5]  # Filter for keywords with at least 5 videos
    top_keywords = keyword_performance.sort_values('mean', ascending=False).head()
    print(top_keywords)
    
    # Category-wise analysis
    print("\nCategory-wise Relevance Scores:")
    category_stats = df.groupby('category')['relevance_score'].agg(['mean', 'median', 'count'])
    print(category_stats)
    
    return df

if __name__ == "__main__":
    df = main()
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
from datetime import datetime, timedelta

# Read the data
df = pd.read_csv('youtube_data/videos_with_relevance.csv')
df['published_date'] = pd.to_datetime(df['published_date'])

# Calculate recency weight (more recent videos get higher weight)
current_date = datetime.now()
df['days_since_published'] = (current_date - df['published_date']).dt.days
df['recency_weight'] = 1 / (1 + df['days_since_published']/365)  # Decay factor over years

def calculate_demand_metrics(keyword_group):
    """Calculate comprehensive demand metrics for a keyword"""
    
    # 1. View Velocity (views per day since publication)
    view_velocity = keyword_group['view_count'] / (keyword_group['days_since_published'] + 1)
    
    # 2. Weighted engagement rate (recent engagement matters more)
    weighted_engagement = (keyword_group['engagement_rate'] * keyword_group['recency_weight']).mean()
    
    # 3. Consistency of demand (standard deviation of view velocity)
    demand_consistency = 1 / (1 + view_velocity.std())  # Normalized to 0-1
    
    # 4. Recent performance (last 6 months)
    recent_mask = keyword_group['days_since_published'] <= 180
    recent_performance = (
        keyword_group[recent_mask]['view_count'].mean() 
        if recent_mask.any() else keyword_group['view_count'].mean()
    )
    
    return pd.Series({
        'avg_daily_views': view_velocity.mean(),
        'weighted_engagement': weighted_engagement,
        'demand_consistency': demand_consistency,
        'recent_performance': recent_performance,
        'video_count': len(keyword_group),
        'total_views': keyword_group['view_count'].sum(),
        'avg_relevance': keyword_group['relevance_score'].mean()
    })

# Calculate demand metrics by keyword
keyword_demand = df.groupby(['category', 'keyword']).apply(calculate_demand_metrics)

# Calculate normalized demand score (0-100 scale)
def normalize_score(series):
    return 100 * (series - series.min()) / (series.max() - series.min())

keyword_demand['demand_score'] = (
    normalize_score(np.log1p(keyword_demand['avg_daily_views'])) * 0.4 +  # View velocity
    normalize_score(keyword_demand['weighted_engagement']) * 0.25 +        # Engagement
    normalize_score(keyword_demand['demand_consistency']) * 0.15 +        # Consistency
    normalize_score(np.log1p(keyword_demand['recent_performance'])) * 0.2 # Recent performance
)

# Calculate supply saturation (inverse of opportunity)
keyword_demand['supply_saturation'] = normalize_score(keyword_demand['video_count'])
keyword_demand['opportunity_score'] = keyword_demand['demand_score'] * (100 - keyword_demand['supply_saturation']) / 100

# Visualize the results
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        'Top Keywords by Demand Score',
        'Demand vs. Supply Analysis',
        'View Velocity Trends',
        'Top Content Opportunities'
    )
)

colors = {'Modern': '#2E7D32', 'Current': '#1565C0', 'Old': '#D84315'}

# Plot 1: Top Keywords by Demand
top_demand = keyword_demand.nlargest(15, 'demand_score')
fig.add_trace(
    go.Bar(
        x=top_demand.index.get_level_values(1),
        y=top_demand['demand_score'],
        marker_color=[colors[cat] for cat in top_demand.index.get_level_values(0)],
        text=[f"{score:.1f}" for score in top_demand['demand_score']],
        textposition='auto',
    ),
    row=1, col=1
)

# Plot 2: Demand vs. Supply Scatter
for category in colors.keys():
    cat_data = keyword_demand.xs(category, level=0)
    fig.add_trace(
        go.Scatter(
            x=cat_data['supply_saturation'],
            y=cat_data['demand_score'],
            mode='markers',
            name=category,
            marker=dict(color=colors[category]),
            text=cat_data.index,
            hovertemplate="Keyword: %{text}<br>Demand: %{y:.1f}<br>Supply: %{x:.1f}"
        ),
        row=1, col=2
    )

# Plot 3: View Velocity Trends
top_velocity = keyword_demand.nlargest(10, 'avg_daily_views')
fig.add_trace(
    go.Bar(
        x=top_velocity.index.get_level_values(1),
        y=top_velocity['avg_daily_views'],
        marker_color=[colors[cat] for cat in top_velocity.index.get_level_values(0)],
        text=[f"{vel:.0f}" for vel in top_velocity['avg_daily_views']],
        textposition='auto',
    ),
    row=2, col=1
)

# Plot 4: Top Opportunities
top_opportunities = keyword_demand.nlargest(15, 'opportunity_score')
fig.add_trace(
    go.Bar(
        x=top_opportunities.index.get_level_values(1),
        y=top_opportunities['opportunity_score'],
        marker_color=[colors[cat] for cat in top_opportunities.index.get_level_values(0)],
        text=[f"{score:.1f}" for score in top_opportunities['opportunity_score']],
        textposition='auto',
    ),
    row=2, col=2
)

# Update layout
fig.update_layout(
    title="Keyword Demand and Opportunity Analysis",
    height=1000,
    width=1200,
    showlegend=True
)

fig.show()

# Print detailed analysis of top opportunities
print("\nTop Content Opportunities Analysis:")
print("-" * 50)
for category in colors.keys():
    print(f"\n{category} Category Top Opportunities:")
    cat_ops = keyword_demand.xs(category, level=0).nlargest(3, 'opportunity_score')
    for keyword, metrics in cat_ops.iterrows():
        print(f"\nKeyword: {keyword}")
        print(f"Demand Score: {metrics['demand_score']:.1f}")
        print(f"Supply Saturation: {metrics['supply_saturation']:.1f}")
        print(f"Opportunity Score: {metrics['opportunity_score']:.1f}")
        print(f"Average Daily Views: {metrics['avg_daily_views']:.1f}")
        print(f"Engagement Rate: {metrics['weighted_engagement']:.2f}%")
        print(f"Demand Consistency: {metrics['demand_consistency']:.2f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import euclidean_distances
from scipy.linalg import eigh
import plotly.graph_objects as go

# Reference the existing data loading and processing
# Lines 3014-3015 for loading the data
# Lines 3900-3941 for the keyword demand metrics

# Prepare the data matrix for PCoA
features = ['demand_score', 'supply_saturation', 'avg_daily_views', 
           'weighted_engagement', 'demand_consistency']
X = keyword_demand[features].values
X_scaled = StandardScaler().fit_transform(X)

# Calculate distance matrix
dist_matrix = euclidean_distances(X_scaled)

# Center the distance matrix
n = dist_matrix.shape[0]
H = np.eye(n) - np.ones((n, n)) / n
B = -0.5 * H.dot(dist_matrix).dot(H)

# Compute eigenvalues and eigenvectors
eigenvals, eigenvecs = eigh(B)

# Sort by eigenvalues in descending order
idx = eigenvals.argsort()[::-1]
eigenvals = eigenvals[idx]
eigenvecs = eigenvecs[:, idx]

# Calculate explained variance ratio
explained_var_ratio = eigenvals / eigenvals.sum()

# Get the first two principal coordinates
pc1 = eigenvecs[:, 0] * np.sqrt(eigenvals[0])
pc2 = eigenvecs[:, 1] * np.sqrt(eigenvals[1])

# Create PCoA plot
fig = go.Figure()

# Add scatter plots for each category
for category in colors.keys():
    mask = keyword_demand.index.get_level_values(0) == category
    
    fig.add_trace(go.Scatter(
        x=pc1[mask],
        y=pc2[mask],
        mode='markers',  # Removed 'text' from mode
        name=category,
        marker=dict(
            color=colors[category],
            size=10,
            opacity=0.7
        ),
        hovertemplate=(
            "Keyword: %{customdata[2]}<br>" +
            "Demand Score: %{customdata[0]:.1f}<br>" +
            "Supply Saturation: %{customdata[1]:.1f}"
        ),
        customdata=np.column_stack((
            keyword_demand[mask]['demand_score'],
            keyword_demand[mask]['supply_saturation'],
            keyword_demand.index.get_level_values(1)[mask]
        ))
    ))

# Update layout
fig.update_layout(
    title="Principal Coordinate Analysis (PCoA) of Keywords",
    xaxis_title=f"PC1 ({explained_var_ratio[0]*100:.1f}% explained variance)",
    yaxis_title=f"PC2 ({explained_var_ratio[1]*100:.1f}% explained variance)",
    height=800,
    width=1000,
    showlegend=True,
    legend=dict(
        title="Categories",
        yanchor="top",
        y=0.99,
        xanchor="right",
        x=0.99
    ),
    plot_bgcolor='white'
)

# Update axes
fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGray', zeroline=True)
fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGray', zeroline=True)

fig.show()

# Print explained variance information
print("\nExplained Variance by Principal Coordinates:")
print(f"PC1: {explained_var_ratio[0]*100:.1f}%")
print(f"PC2: {explained_var_ratio[1]*100:.1f}%")
print(f"Total: {(explained_var_ratio[0] + explained_var_ratio[1])*100:.1f}%")
```
